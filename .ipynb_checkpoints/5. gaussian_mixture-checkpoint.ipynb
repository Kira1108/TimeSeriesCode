### Probability Distribution of Gauss Mixtures

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

x_list = np.linspace(-0.1, 0.1, 500)

p = 0.5

fx = p * stats.norm.pdf(x_list, 0, 0.1) + (1 - p) * stats.norm.pdf(x_list, 0, 0.02)

plt.plot(x_list, fx)
plt.show()
```

```python
import seaborn as sns

samples = []

m0, s0 = 0,0.01
m1, s1 = 0,0.002

from scipy import stats
p = 0.5

d1 = stats.norm.rvs(loc=m0, scale=s0, size=5000)
d2 = stats.norm.rvs(loc=m1, scale=s1, size=5000)
mask = (np.random.random(size=5000) < p)
mixture = mask * d1 + (1 - mask) * d2

sns.kdeplot(mixture, shade=True)
plt.title("Gaussian Mixture Sample")
plt.grid(linestyle = '--')
plt.show()
```

```python
print("Kurtosis:", stats.kurtosis(mixture))
```

### Gaussian mixture model for sbux log return

```python
from sklearn.mixture import GaussianMixture
from fintools import calculate_return
import pandas as pd

file_path = "data_files/financial_basis/all_stocks_5yr.csv"
df = pd.read_csv(file_path, parse_dates = ['date'])

def select_stock(stock:str):
    return df[df['Name'] == stock].copy().pipe(calculate_return)

sbux = select_stock("SBUX")
```

```python
data = sbux['log_return'].dropna().values.reshape(-1,1)
model = GaussianMixture(n_components=2)
model.fit(data)
```

```python
weights = model.weights_
means = model.means_
cov = model.covariances_
```
