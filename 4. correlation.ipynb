```python
%config InlineBackend.figure_format = 'retina'

filepath = "./data_files/time_series_analysis/sp500_close.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
close = pd.read_csv(filepath)
```

```python
import seaborn as sns

sns.heatmap(
    close[['AAPL','ADI',"XL"]]\
        .dropna(axis = 0, how = 'all')\
        .pct_change()\
        .dropna(axis = 0, how = 'all')\
        .corr(),
    annot = True,
)
plt.show()
```
